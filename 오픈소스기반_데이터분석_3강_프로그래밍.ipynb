```python
# 실습1
def get_dataset_stats(values):
  return min(values), max(values), sum(values) / len(values)

data = [12, 8, 21, 17, 5, 32, 14]

# 함수값 return 언패킹
minimum,maximum,average=get_dataset_stats(data)
print(f'최솟값: {minimum}, 최댓값: {maximum}, 평균: {average:.2f}')
```

```python
# map 함수 : 반복 가능한 객체의 모든 요소에 특정 함수를 적용하여 새로운 값을 생성하는 함수 / 리스트나 튜플과 같은 자료구조의 요소를 변경할 때 유용
# map(a,b) : a=통상 함수 형태,  b=값을 변경시킬 대상의 변수 / b 변수에 대해 a를 적용한 값을 소환
temperature_celsius = [25.6, 27.8, 30.5, 22.3, 28.9,31.2, 24.7]
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32
temperature_fahrenheit = list(map(celsius_to_fahrenheit, temperature_celsius)) # 위에서 변환한 값을 temperature_celisus에 넣어서 값을 변환시킴
print("섭씨 온도 데이터:", temperature_celsius)
print("화씨 온도로 변환:", temperature_fahrenheit)
```

```python
# filter 함수 사용하기
numbers=[10,15,20,25,30]
even_numbers=list(filter(lambda x: x%2==0,numbers)) # filter(a,b) : b 변수에 대해 a 조건에서 True인 것만 소환
print(even_numbers)
```

```python
# filter 함수 사용하기
osda_students = [
{"name": "김지원", "score": 85},
{"name": "이민준", "score": 65},
{"name": "박서연", "score": 90},
{"name": "정현우", "score": 55},
{"name": "최예은", "score": 78} ]
passed_students = list(filter(lambda student:student["score"] >= 70, osda_students)) # 해당 lambda 함수 조건에 해당하는 것만 기존 변수에서 소환 또는 남
print(passed_students,"\n")

print("70점 이상 합격한 학생 목록:")
for student in passed_students:
  print(f"이름: {student['name']}, 점수: {student['score']}")
```

```python
# reducce : 반복 가능한 객체의 모든 요소를 누적하여 하나의 값으로 축소하는 함수 / 반복문 필요 없음 / functtool 모듈에서 제공하는 함수
from functools import reduce
numbers = [1, 2, 3, 4, 5]
total_sum = reduce(lambda x, y: x + y, numbers)
print(total_sum)
```

```python
numbers=[3,7,2,9,5]
max_value=reduce(lambda x,y: x if x>y else y,numbers)
# 3, 7 소환해서 7이 선택 / 그러면 [7,9,5]가 남은 / 7,9 비교해서 9가 남음. 그리고 [9,5]가 남음
print(max_value)
```

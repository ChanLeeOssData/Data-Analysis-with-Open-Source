# 리스트 컴프리헨션
image.png

```python
# 기존 방법으로 리스트 생성
numbers = []
for i in range(10):
  numbers.append(i)
print(numbers)
```

```python
# 리스트 컴프리헨션으로 리스트 생성
numbers = [ i for i in range(10) ]
print(numbers)
```

```python
# 짝수만 추출하기 + 기본 코드
even_num=[]
for i in numbers:
  if i%2==0:
    even_num.append(i)
print(even_num)
```

# 튜플 컴프리헨션

```python
list1=['사과','복숭아','바나나']
list2=['주스','잼','통조림']
pairs=[]

# 두 리스트에 포함된 모든 원소를 조합한 투플의 리스트 생성하기 + 기본 코드
for fruit in list1:
  for product in list2:
    pairs.append((fruit,product))
print(pairs)
```

# 딕셔너리 컴프리헨션
image.png

```python
# 딕셔너리로 생성하기
# 복잡하게 코딩하기
squares={}
for i in range(5):
  squares[i]=i**2
print(squares)
```

```python
# 딕셔너리 컴프리헨션으로 딕셔너리 생성하기
print({i:i**2 for i in range(5)})
```

```python
city_population={"서울":957,"부산":339,"인천":294,"대구":242,"광주":145}

# 인구 200만명 이상의 도시 리스트
print({city:pop for city,pop in city_population.items() if pop>=200})
```

```python
# 인구 300만명 이상 및 이름에 "산"이 포함된 도시 리스트
print({city:pop for city,pop in city_population.items() if pop>=300 and '산' in city})
```

## {:.2f} - 소수점 둘째자리까지    /  {a:.2f} - f 문자열에서의 방법임
## {:,} - 1000자리 구분 쉽표
## {:d} - 정수 표시
## {:10} - 우측으로 10칸 정렬
## {:<10} - 좌측으로 10칸 정렬
## {:^10} - 중앙으로 10칸 정렬

```python
name="홍길동"
age=30
salary=3500000
tax_rate=0.1

# C 스타일 문자열 포매팅
print('이름: %s, 나이: %d, 월급: %d원' % (name,age,salary))
```

```python
# format 메소드 방법
print('이름: {}, 나이: {}세, 월급: {}원'.format(name,age,salary))

print('이름: {0}, 나이: {1}세, 월급: {2}원, 매 월 받는 돈: {2}원'.format(name,age,salary))
```
